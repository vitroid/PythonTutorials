Make sample file for regression.

```python
%matplotlib inline
import numpy as np

t    = np.linspace(0,5,num=1000)
data = np.random.randn(*t.shape)*3-4*t+4
import pylab
pylab.plot(t, data)
file = open("sample3.txt", "w")
for x,d in zip(t,data):
    file.write("{0} {1}\n".format(x,d))
```

```python
xsum = 0.0
ysum = 0.0
xxsum = 0.0
xysum = 0.0
n   = 0.0

file = open("sample3short.txt")
for line in file:
    x,y = [float(v) for v in line.split()]
    xsum += x
    ysum += y
    xxsum += x**2
    xysum += x*y
    n   += 1.0

D = 1.0 / (n*xxsum - xsum**2)
a = n*xysum - xsum*ysum
b = -xsum*xysum + xxsum*ysum
a *= D
b *= D
c = xysum / xxsum
print(a,b,c)
```
