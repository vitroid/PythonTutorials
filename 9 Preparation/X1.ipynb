Make sample file for averaging.

```python
%matplotlib inline
import numpy as np

t    = np.linspace(0,5,num=1000)
data = np.random.randn(*t.shape)*2.5+10
import pylab
pylab.plot(data)
file = open("sample2.txt", "w")
for d in data:
    file.write("{0}\n".format(d))
```

```python
file = open("sample2short.txt")
n = 0.0
sum = 0.0
ssum = 0.0
for line in file:
    value = float(line)
    n += 1
    sum += value
    ssum += value**2
sum /= n
ssum /= n
sig2 = ssum - sum**2
precision = (sig2*n/(n-1))**0.5
error     = precision / n**0.5
print("平均: {0}".format(sum))
print("分散: {0}".format(sig2))
print("精度: {0}".format(precision))
print("誤差: {0}".format(error))
```
