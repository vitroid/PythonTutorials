# はじめに
実験データの誤差の取り扱いについて少し話します。そのあと、データ処理プログラムを作ってもらって、サンプルデータで実際にうまくいくかどうか確認します。

## 期待値
体温計で体温を測るときに、何回か繰り返して測ったことはあるでしょうか。意外に同じ数字は出ないものです。こんな風に、毎回違った測定値が出るのは、誤差が大きいのでしょうか? それとも精度が悪いのでしょうか? と尋ねられた時、即答できる人はあまりいないのではないかと思います。精度とか誤差って何なんでしょう。

体温は時々刻々変化している可能性があるので、測るたびに変化するのはやむをえないかもしれません。そこで、もう少し変化しそうにないもの、例えば缶コーヒーの缶の直径をものさしで測ることにします。

ある講義で、定規と缶を学生さんに順番に回して、同じ定規と缶の組みあわせで直径を測ってもらったことがあるのですが、丸いものは意外に測りにくいのか、人によって2 mmぐらいの差がでてきます。10人に測ってもらった時の測定結果は次のようになりました。

> 51 51 50 50 50 51 50 50 49 50 (単位はmm)

このように、測定する道具(ものさし)の都合で、測定値がばらつく時、そのばらつきの大きさを*精度*と呼びます。直径を測るのにもっと適した道具、たとえばノギス(下図)を使えば、より高い*精度*で測ることができます。つまり、*精度はものさしに固有の性質であって、缶の性質ではありません。*

![Caliper](https://upload.wikimedia.org/wikipedia/commons/thumb/f/f6/Vernier_caliper.svg/500px-Vernier_caliper.svg.png)

では、このものさしの測定値から、缶の直径はいくらだと言えるでしょうか。データはばらついているものの、あるデータがほかのデータより特に間違っているという確かな証拠はないので、どのデータも同じぐらい信用するものとしましょう。そして、公平を期するために、 個々の測定値 $x_i$ と$X$のずれ(偏差)が、最も小さくなるように、缶の直径の期待値$X$を選びます。

以後、個々の測定値を $x_i$、測定値全体を$\{x_i\}$と書くことにします。

偏差は正になることも負になることもあるので、偏差の総量は二乗和であらわすことにし(脚注1)、これを$I$と書くことにすると、
$$I=\sum_{i=1}^{10}(x_i-X)^2$$
と書けます。$I$が極小になる条件、つまり${\rm d}I/{\rm d}X = 0$をみたす$X$を求めると、
$$\frac{{\rm d}I}{{\rm d}X}=2\sum_{i=1}^{10}(X-x_i)$$
$$=20X-2\sum_{i=1}^{10}x_i=0$$
$$X=\frac{1}{10}\sum_{i=1}^{10}x_i$$
何のことはない、缶の直径の期待値$X$は、缶の測定値の(相加)平均に等しいことがわかります。あたりまえ?  いえ、そんなことはありません。缶の直径の期待値をどうやって決めるかは、自明なことではないからです。今回は偏差$x_i-X$を小さくするという方針を採用したので、期待値は相加平均と等しくなりましたが、そうじゃない方針を選ぶ自由もあります。相乗平均のほうが適切な場合もあります。特に、個々のデータの信頼性にばらつきがある場合には、単純な平均よりももっとうまく期待値を決める方法があるはずです。期待値を計算する時には、本当に単純な平均をとって良いのか、一歩立ち止まって考えてみて下さい。


## 精度
さて、缶の直径の期待値は、測定値の平均 50.2mm と定まりました。これをもとに、測定値のばらつき=精度を求めましょう。

ばらつきの大きさを数値化するのに、期待値を基準にして、そこからのずれ (つまり偏差) を使う方法と、期待値はとりあえず考えず、測定値同士の差だけを使う方法が考えられます。どちらでもいいのですが、今は後者の方針をとってみることにしましょう。(ここでも、あなたは別の方針を選ぶ自由があります)

10個の測定値から2個を選んできて、その差を二乗したものを$a_{ij} = ( x_i – x_j )^2$と書きます。二乗にするのは、2つの測定値のうちどちらが大きいかを条件分けするのが面倒だからです。

測定値が10個あれば、その中から2個選ぶ組み合わせは全部で45通りありますから、45通りすべての組み合わせでの$a_{ij}$の平均を求めてみましょう。
$$\frac{1}{45}\sum_{i=1}^{10}\sum_{j=i+1}^{10}a_{ij}
=\frac{1}{45}\left[\sum_{i=1}^{10}\sum_{j=1}^{10}a_{ij}\right]/2$$
$$=\frac{1}{90}\sum_{i=1}^{10}\sum_{j=1}^{10}(x_i^2+x_j^2 - 2x_ix_j)$$
$$=\frac{1}{90}\left[2\sum_{i=1}^{10}10x_i^2-2\sum_{i=1}^{10}x_i\sum_{j=1}^{10}x_j\right]$$
$$=\frac{1}{90}\left[20\sum_{i=1}^{10}x_i^2-200X^2\right]$$
$$=\frac{200}{90}\left[\sum_{i=1}^{10}x_i^2/10-X^2\right]$$

角カッコの中の第1項は、測定値の二乗和を個数で割ったもので、**二乗平均**と呼ばれます。第2項は平均の二乗です。そして、二乗平均から平均の二乗を引いたものは、**分散**と呼ばれ、記号$\sigma^2$で表します。分散の平方根$\sigma$は**標準偏差**と呼びます。角カッコの前についている分数は、データ数を$N$とすると、分子が$2N^2$、分母が$N(N–1)$ですから、約分すると$2N/(N–1)$となります。

$a_{ij}$の平均は求められましたが、これを使って精度をどう定義するかはまだ任意性があります。ここでは、$a_{ij}$の平均の半分の平方根を、**精度**と定義することにします。「半分」が出てくるのが実に気持悪いですが、こう定義すると、測定値の個数$N$が非常に大きい場合に、精度が測定値の標準偏差$\sigma$に等しくなり、何かと都合が良いのです。

こうして定義した精度$p$を式で書くと、
$$p=\sigma\sqrt{\frac{N}{N-1}}$$
となります。分母に$N-1$があることから、データが1個しかない場合は精度が定義できないことを示しています。また、$N$が十分大きくなると、$p$は$\sigma$に等しくなります。上の測定値の場合、標準偏差が0.6 mmなので、ものさしの精度は0.7 mmとなります。

期待値を計算してみたら、それが測定量の平均になったのと同じように、精度を計算してみたら、それは測定量の標準偏差に近い値になりました。しかし、期待値と平均の関係、精度と標準偏差の関係は、何を期待値と定義するか、何を精度と定義するかに依存しているということを、ちょっと心に留めておいて下さい。



## 誤差
ここまでの説明で誤差は一度も現れませんでした。では、誤差とは何でしょう。

誤差とは、期待値$X$の不確かさ、つまり測定値から予測される、缶の直径の不確かさのことです。それは精度と同じじゃないか、と思うかもしれませんが、違います。考えてみて下さい。10個のばらついた測定値から、その期待値Xをまず決めました。次に、もう一度測定をします。11個目の測定データは、やはり期待値にはぴったり一致しないので、偏差が生じますが、11番目の測定値を含めて計算しても、期待値$X$はほんのちょっとしか変化しません。測定データが100万個ぐらいになると、期待値$X$は多少測定値が増えても、びくとも変化しなくなります。個々の測定は、ものさしの性能のせいで、精度$p$程度の幅でばらつくとしても、期待値$X$は測定値が多いと、ほとんど変動しなくなります。つまり、期待値の不確かさ = 誤差は測定回数が増えるほど小さくなるのです。

正確に言うと、誤差$e$は測定回数の平方根に反比例して減少します。式で書くと、
$$e=\sigma/\sqrt{N-1}=p/\sqrt{N}$$
となります。 データが1個しかなければ、精度$p$が計算できないので、誤差も不明です。

繰り返しますが、誤差は期待値(平均値)の不確かさを表しています。精度は、ものさしの不確かさを表しています。 精度は、測定回数には関係なく一定です。 精度の低いものさしではかると、誤差は大きくなりますが、何度も測定すれば誤差は小さくすることができます。誤差を1桁小さくするためには、測定を100倍繰り返す必要があります。
### 誤差の性質(1)
測定値を10倍すると、誤差、精度とも10倍になります。これは、センチメートル単位で測定した量を、ミリメートルに書きなおすケースを想像すればわかるでしょう。誤差や精度の単位は測定値の単位と同じですから、測定値の単位をとりなおすと誤差の単位も変わります。測定値の分散(精度の二乗)は100倍になります。

測定値に定数$c$を加えると、平均値も$c$だけ増えますが、分散は変化しないので、精度、誤差とも変化しません。
### 誤差の性質(2)
別々に測定した2つの量$x$と$y$の和を考えてみます。(例えば、200 ccの計量カップで、400 ccの水を量る、はがきのたてとよこの長さの和をとる、バイクの前輪と後輪の荷重から総重量を推定する、など)測定値$\{x_i\}$は全部で$M$個、測定値$\{y_i\}$は全部で$N$個あるものとします。

$x$と$y$は独立に測定したデータですから、それらの和をとる方法は、全部で$M\times N$通りあります。そこで、$M\times N$通りの組み合わせを使って、$z = x + y$の期待値と、その分散を計算します。

$z$の期待値$Z$は、
$$Z=\frac{1}{MN}\sum_{i=1}^M\sum_{j=1}^N(x_i+y_j)$$
$$Z=\frac{1}{MN}\sum_{i=1}^M\left(Nx_i+\sum_{j=1}^Ny_j\right)$$
$$=X+Y$$
つまり、それぞれのデータの平均値同士を足しあわせれば、和$z$の期待値が得られます。

個々の和$x_i + y_j$は、期待値$X + Y$からは$x_i + y_j - ( X + Y )$だけずれていますから、このずれ(偏差)の二乗の総和をとり、組合せの数$MN$で割ると、二乗偏差の期待値、つまりzの分散が求まります。計算が面倒なので、まず個々のデータの偏差を次のように定義します。
$$\delta x_i = x_i – X$$
$\delta x_i$の平均は0、$\delta x_i$の分散は$x_i$の分散と同じです。これを使うと、式が簡単にできて、
$$\sigma_z=\frac{1}{MN}\sum_{i=1}^M\sum_{j=1}^N(x_i+y_j-(X+Y))^2$$
$$=\frac{1}{MN}\sum_{i=1}^M\sum_{j=1}^N(\delta x_i+\delta y_j)^2$$
$$=\frac{1}{MN}\sum_{i=1}^M\left(N\delta x_i^2+\sum_{j=1}^N(\delta y_j^2+2\delta x_i\delta y_j)\right)$$
$x_i$と$y_j$は独立な測定量なので、共分散
$$\frac{1}{MN}\sum_{i=1}^M\sum_{j=1}^N\delta x_i\delta y_j$$
は0になります。結果として、二乗偏差の期待値は
$$\sigma_z^2=\frac{1}{MN}\sum_{i=1}^M\left(N\delta x_i^2+\sum_{j=1}^N\delta y_j^2\right)$$
$$=\sigma_x^2+\sigma_y^2$$
となり、個々のデータの分散の和になりました。上の式をじっくり見るとわかりますが、$x + y$ ではなく、$x - y$の二乗偏差の期待値を計算する場合も、共分散が0になることから、$x - y$ の二乗偏差の期待値は、$x$ と $y$ それぞれの**分散の和**になります。

データ数が多ければ、その標準偏差$σ$は、ほぼ精度に等しいことは前節で示しました。上の考察から、実際にすべてのデータの組合せを計算しなくても、和と差の精度は個々のデータの精度から直ちに計算できることがわかります。また、 2つのデータの和や差の分散が、個々のデータの分散の和になるということは、和や差の精度は、もとのデータの精度よりも、必ず悪くなることを意味します。

$z$の誤差はどうなるでしょうか。$z$の誤差とは、$z$の期待値の不確かさのことです。前節で説明した通り、 $x$の誤差は、$x$の測定回数$M$が増えるほど、その平方根に逆比例して小さくなります。もし、$y$のデータが極めて正確な場合には、$z$の誤差$e_z$は$x$の誤差$e_x$に等しいでしょう。同じことは、$y$についても言えます。ですから、$z$の誤差は次の性質を満たすはずです。

1. $y$の誤差が0なら、$z$の誤差は$x$の誤差に等しい。
2. $x$の誤差は、$x$の標準偏差を$x$のデータ数$M$の平方根で割ったものである。

これらを合せて考えると、$z$の誤差$e_z$は
$$e_z^2 = e_x^2 + e_y^2$$
と書くのが合理的でしょう。ただし、$e_x^2 = \sigma_x^2 / M, e_y^2 = \sigma_y^2 / N$ です。 誤差についても、精度と同じ形の関係式がなりたち、測定値同士の和や差をとると、必ず誤差が大きくなります。

このような性質のことを、「**誤差の伝搬法則**」と呼びます。
### 誤差の性質(3)
測定データ$\{x_i\}$ (平均$X$、分散$\sigma_x^2$)を、ある関数 $y = f(x)$に入れた場合の、$y$の平均と分散を計算してみます。ここでは、取り扱いをとても簡単にするために、データ $x_i$の範囲で、おおよそ$f(x)\sim a x + b$と書けるものと仮定します。すると、誤差の性質(1)から、$y$の平均は$Y = a X + b = f(X)$、分散は$\sigma_y^2 = a^2 \sigma_x^2 = (f’(X))^2\sigma_x^2$ となることがわかります。ただし、$f’(x)$ は、関数 $f(X)$ の勾配(導関数)です。

これを誤差の性質(2)と組みあわせると、2つの測定データの積の平均と分散が計算できます。

1. 2つの測定データセットを、$\{x_i\}$ (平均$X$、分散$\sigma_x^2$) と$\{y_i\}$(平均$Y$、分散$\sigma_y^2$)とします。(例えば、はがきの縦と横の長さ) 
2. これらの対数を$p_i =\ln x_i、 q_i = \ln y_i$ と書くことにすると、$\{p_i\}$の平均 $P$ と分散$\sigma_p^2$は、誤差の性質(3)より、それぞれ$P = \ln X, \sigma_p^2 = (f’(X))2 \sigma_x^2 =\sigma_x^2  / X^2$ となります。$Q$も同様です。
3. $p_i$ と $q_i$ の和を $r_i$ とすると、$\{r_i\}$ の平均$R$と分散$\sigma_r^2$は、誤差の性質(2)より、それぞれ$R = P + Q, \sigma_r^2 = \sigma_p^2 + \sigma_q^2$となります。
4. 最後に、$z_i = \exp r_i$ と定義します。$\{z_i\}$ の平均 $Z$ と分散$\sigma_z^2$は、誤差の性質(3)より、それぞれ$Z = \exp R, \sigma_z^2 = (f’(R))^2 \sigma_r^2 = (f(R))^2 \sigma_r^2 = Z^2 \sigma_r^2$となります。
5. $z_i = \exp r_i = \exp( p_i + q_i ) = \exp( \ln x_i + \ln y_i ) = \exp( \ln x_i y_i ) = x_i y_i,\\
Z = \exp R = \exp( P + Q ) = \exp( \ln X + Y ) = \exp( \ln X Y ) = X Y,\\
\sigma_z^2 / Z^2 = \sigma_r^2 = \sigma_p^2 + \sigma_q^2 = \sigma_x^2  / X^2 + \sigma_y^2  / Y^2$

標準偏差$\sigma$(つまり精度)を平均で割ったものを相対精度と呼ぶことにすると、計測量同士を掛けあわせた場合には、相対精度の二乗の可成則がなりたつことがわかります。
### 誤差の表記
測定値の誤差を表記する方法には3通りあります。

1. $l = (a\pm b)$ cm あるいは$l = a$ cm $\pm b$ cm
2. $l = a (c)$ cm					
3. $l = a_b$ cm						

#### 例

1. $l = (5.3478 \pm 0.0065)$ cm, $l = 5.3478$ cm $\pm 0.0065$ cm
2. $l = 5.3478(32)$ cm
3. $l = 5.34_8$ cm

(1)のように表した場合は、「長さ $l$ の真の値が、不確かさの範囲 ( $\pm b$ ) に含まれる確率が極めて高い」という意味です。誤差(測定値の期待値の標準偏差)を$\sigma$とすると、$b\gt 2\sigma$ とするように勧告されていますが(脚注2)、論文によっては $b = \sigma$ としている場合もあり、注意が必要です。

(2)のように表した場合は、「カッコの中の数値 $c$ は、$a$の最後の桁に適用され、$1\sigma$を表す」という意味になります。したがって、例では期待値5.3478で標準偏差は0.0032と推定されます。

(3)の表記は不確かさが精密に推定できない場合に用いられ、$a$ の不確かさの推定値は下付き添字$b$の数字の桁で、1と9の間にあるとみなされます。例では$l\sim 5.348$ cmで、最後の桁の8の数字の不確かさは精密に推定できないが、1以上で9以下であると解釈されます。つまり、最悪の場合、不確かさは9なので、$l$ の数値は5.339 cmから5.357 cmの間に入ると解釈されます。
### 有効数字
電卓やパソコンを使って期待値や誤差を計算すると、期待値$\mu = 5.3478329874$、誤差$\sigma = 0.003673648$ といったように、大きな桁数の数字が得られますが、この数字を最終的にレポートに記載する時には 5.347832987(3673648) と表すべきか、5.348(4) で良いのか、あるいは 5.3478(37) と表すべきでしょうか。
これを判断するためには、誤差の不確かさがどの程度になるか、つまり「誤差の精度」を考える必要があります。細かい議論は省きますが、$N$個の測定値から誤差を推定する場合、誤差の確からしさは誤差そのものの$1/\sqrt{N}$程度しかありません。つまり、100 個の測定値から誤差が0.003673648と求まったとすれば、実際には誤差の値には 10% の不確かさ、つまり 0.0033〜0.0040 ぐらいの不確かさがあり、1桁しか信頼性がありません。そこで、誤差の 2 桁目の数字を切り上げて(脚注3) 、 5.348(4) と表記するのが適当です。前節の例(1)(2)のように、誤差を 2 桁の数値で表すためには、1,000 回〜10,000 回程度測定を繰り返す必要があります。


## プログラム課題1

実験値が書かれたサンプルデータsample2short.txtを読みこんで、データの個数と期待値と精度と誤差を表示するプログラムを作ってください。sample2short.txtには4つしかデータがありません。それらは10、12、10、8ですから、平均は10、分散$\sigma^2$は2、精度は1.6、誤差は0.8のはずです。

sample2short.txtでの計算がうまくいったら、sample2.txtを読み込むように書き換えて下さい。

分散は次の式で定義されます。
$$\sigma^2={1\over N}\sum_i \left(x_i-X\right)^2$$
また、これを変形して次のようにも書けます。
$$\sigma^2={1\over N}\sum_i x_i^2 - \left({1\over N}\sum_i x_i \right)^2$$
