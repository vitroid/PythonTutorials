# プロットの基本

* x, yそれぞれの系列のデータを準備する。

```python
x = [1,2,3,4,5]
y = [2,4,8,16,32]
# データの数は同じにすること。
```

```python
from matplotlib import pyplot as plt

plt.plot(x, y)
```

線を変える、色を変える、幅を変える。

```python
plt.plot(x, y, color="red", linestyle="dashed", linewidth=5)
```

もっと簡易な指定方法。(r=red, o=circular markers, -- = dashed line)

```python
plt.plot(x, y, "ro--")
```

x軸、y軸にラベルをつけ、範囲指定し、さらに凡例を付ける。

```python
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.xlim(0,6)
y2 = [13,8,5,3,2]
plt.plot(x, y, color="red", linestyle="dashed", linewidth=5, label="Data 1")
plt.plot(x, y2, color="blue", linestyle="solid", linewidth=3, label="Data 2")
plt.legend()
```

FigureをPDFファイル化する。

```python
fig = plt.figure() # 最初に
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.xlim(0,6)
y2 = [13,8,5,3,2]
plt.plot(x, y, color="red", linestyle="dashed", linewidth=5, label="Data 1")
plt.plot(x, y2, color="blue", linestyle="solid", linewidth=3, label="Data 2")
plt.legend()
fig.savefig("data12.pdf") # 最後に

#PDFファイルは画面左端のフォルダーアイコンからダウンロードできます。
```

データがx,yのペアになっている場合はこんな風に変換します。

```python
pairs = [[1,1],
         [2,2],
         [3,4],
         [4,8],
         [5,16]]

x = [pair[0] for pair in pairs]
y = [pair[1] for pair in pairs]
plt.plot(x,y)
```
