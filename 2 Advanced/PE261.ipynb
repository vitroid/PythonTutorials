```python
def sqsum(k,m):
    total = 0
    for i in range(k-m,k+1):
        total += i**2
    return total

def right(total,m):
    n = (total/m)**0.5 + m/2
    n = int(n)
    t = sqsum(n,m-1)
    if t == total:
        return n-m+1
    if t > total:
        while t > total:
            n -= 1
            t = sqsum(n,m-1)
            if t == total:
                return n-m+1
        return -1
    while t < total:
        n += 1
        t = sqsum(n,m-1)
        if t == total:
            return n-m+1
    return -1

    

sqsum(24,4)
```

```python
right(2430,4)
```

上は単純に機械的に計算した。右辺と左辺をキャンセルすることで、もっと良い計算ができると思う。
$$(k-m)^2+\cdots+k^2=(n+1)^2+\cdots+(n+m)^2$$
$$(n+m)^2-(k-m)^2+\cdots+(n+1)^2-(k-1)^2=k^2$$
$$(n+k)(n-k+2m)+\cdots+(n+k)(n-k+2)=k^2$$
つまり、$k^2$は$n+k$の倍数である。
$$2+4+\cdots+2m=m(m+1)$$
$$(n+k)m(n-k+m+1)=k^2$$
$m$の倍数でもある。

$k$と$m$を決めると、$n$は一意に決まる。

1. $k^2$は$m$で割り切れなければいけない。
2. $D=4k^2 - 4km - 4k + 4k^2/m + m^2 + 2m + 1$が平方数でなければいけない。
3. $n=(-m+\sqrt{D}-1)/2$である。

```python
def issq(n):
    m = n**0.5
    return m == int(m)

def pe261(k):
    for m in range(1,k):
        if k**2 % m != 0:
            continue
        km=k**2/m
        D = 4*k**2 - 4*k*m - 4*k + 4*km + m**2 + 2*m + 1
        n = (-m+D**0.5-1)/2
        if n < k:
            break
        if not issq(D):
            continue
        yield m,n

for k in range(1,10000):
    for m,n in pe261(k):
        print(k,m,n)
```

### 初項

初項は$m=1$の場合、つまり左辺に2乗項が2つ、右辺に1つの、ピタゴラス型の式で、しかも左辺の2項が1違いの場合。
$$(n+k)(n-k+2)=k^2$$
となるようなケースである。これは解ける。

$$n^2+2n+2k(1-k)=0$$

$k$に対し、$n=\sqrt{2k^2-2k+1}-1$である。
