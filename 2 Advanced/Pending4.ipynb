```python
# 英単語辞書の読みこみ
file = open("CT_voc.csv")
words = set()
for line in file:
    cols = line.split(",")
    words.add(cols[1].upper())

# 文字の置きかえ 16進数で表現できる単語だけ抽出
hexwords = set(["CAFE","BABE","DECAF","DECAFED","BEEF","D1ED","ADDED","BEAD","DEC1DED","FACED"])
import re
for word in words:
    result = word.translate(str.maketrans('IO','10'))
    result2 = re.match(r"^[0-9A-F]*$", result)
    if result2 is not None:
        hexwords.add(result)
hexwords
```

```python
import sympy

for w in hexwords:
    i = eval("0x"+w)
    if len(sympy.factorint(i)) == 1:
        print(w,i)
```

```python
# 2-word を生成
import itertools as it
import sympy

phrases = []
for phrase in it.product(hexwords, repeat=2):
    s = "".join(phrase)
    i = eval("0x"+s)
    if len(sympy.factorint(i)) == 1:
        phrases.append(" ".join(phrase))
sorted(phrases, key=lambda x:len(x))
```

```python
# 3-word を生成
import itertools as it
import sympy

with open("3word.txt","w") as file:
    for phrase in it.product(hexwords, repeat=3):
        s = "".join(phrase)
        i = eval("0x"+s)
        if len(sympy.factorint(i)) == 1:
            file.write(" ".join(phrase)+"\n")
```
