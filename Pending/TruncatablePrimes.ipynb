右に延ばしている場合

```python
from sympy.ntheory.primetest import isprime
queue = []

def CutOffPrime(n, digits):
    for i in range(1,10):
        x = i*10**digits + n
        if isprime(x):
            queue.append([x, digits+1])

CutOffPrime(0,0)
while len(queue) >0:
    x, d = queue.pop(0)
    print(x,d)
    CutOffPrime(x,d)
```
